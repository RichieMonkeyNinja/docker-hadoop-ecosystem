```python
def compare_df_dtypes(df1, df2, df3, names=["DF1", "DF2", "DF3"]):
    # Union of all columns
    all_cols = sorted(set(df1.columns) | set(df2.columns) | set(df3.columns))

    dtype_comparison = pd.DataFrame({
        names[0]: df1.reindex(columns=all_cols).dtypes.astype(str),
        names[1]: df2.reindex(columns=all_cols).dtypes.astype(str),
        names[2]: df3.reindex(columns=all_cols).dtypes.astype(str),
    })

    # Flag rows where not all dtypes are identical
    differences = dtype_comparison[
        dtype_comparison.nunique(axis=1) > 1
    ]

    if differences.empty:
        print("✅ All columns have matching dtypes across all DataFrames.")
        return None
    else:
        print("❌ Found dtype differences in the following columns:")
        return differences

important_dtypes = {
    'host_listings_count': 'float64',
    'host_total_listings_count': 'float64',
    'license':'str',
    'minimum_minimum_nights':'float64',
    'maximum_minimum_nights':'float64',
    'minimum_maximum_nights':'float64',
    'maximum_maximum_nights':'float64'
}

df_bristol_2025_09_26_listings = df_bristol_2025_09_26_listings.astype({k: v for k, v in important_dtypes.items() if k in df_bristol_2025_09_26_listings.columns})
df_bristol_2025_06_24_listings = df_bristol_2025_06_24_listings.astype({k: v for k, v in important_dtypes.items() if k in df_bristol_2025_09_26_listings.columns})
df_bristol_2025_03_19_listings = df_bristol_2025_03_19_listings.astype({k: v for k, v in important_dtypes.items() if k in df_bristol_2025_09_26_listings.columns})

df_edinburgh_2025_09_21_listings = df_edinburgh_2025_09_21_listings.astype({k: v for k, v in important_dtypes.items() if k in df_edinburgh_2025_09_21_listings.columns})
df_edinburgh_2025_06_15_listings = df_edinburgh_2025_06_15_listings.astype({k: v for k, v in important_dtypes.items() if k in df_edinburgh_2025_09_21_listings.columns})
df_edinburgh_2025_03_08_listings = df_edinburgh_2025_03_08_listings.astype({k: v for k, v in important_dtypes.items() if k in df_edinburgh_2025_09_21_listings.columns})

df_london_2025_09_14_listings = df_london_2025_09_14_listings.astype({k: v for k, v in important_dtypes.items() if k in df_london_2025_09_14_listings.columns})
df_london_2025_06_10_listings = df_london_2025_06_10_listings.astype({k: v for k, v in important_dtypes.items() if k in df_london_2025_09_14_listings.columns})
df_london_2025_03_04_listings = df_london_2025_03_04_listings.astype({k: v for k, v in important_dtypes.items() if k in df_london_2025_09_14_listings.columns})
```

```python
gdf_listings = gpd.GeoDataFrame(
    df_bristol_2025_09_26_listings,
    geometry=gpd.points_from_xy(
        df_bristol_2025_09_26_listings.longitude,
        df_bristol_2025_09_26_listings.latitude
    ),
    crs="EPSG:4326"
)
```

```python
gdf_joined = gpd.sjoin(
    gdf_listings,
    df_bristol_2025_09_26_neighbourhoods[["neighbourhood", "geometry"]],
    how="inner",
    predicate="within"
)
```

```python
gdf_neighbourhoods_metric = df_bristol_2025_09_26_neighbourhoods.to_crs(epsg=3395)
```

```python
df_bristol_2025_09_26_neighbourhoods["area_sqkm"] = (
    gdf_neighbourhoods_metric.geometry.area / 1e6
)
```

```python
df_bristol_2025_09_26_neighbourhoods["centroid"] = df_bristol_2025_09_26_neighbourhoods.geometry.centroid
df_bristol_2025_09_26_neighbourhoods["centroid_lat"] = df_bristol_2025_09_26_neighbourhoods.centroid.y
df_bristol_2025_09_26_neighbourhoods["centroid_lon"] = df_bristol_2025_09_26_neighbourhoods.centroid.x

# Compactness: perimeter² / (4π × area)
df_bristol_2025_09_26_neighbourhoods["perimeter_m"] = (
    gdf_neighbourhoods_metric.geometry.length
)
```

```python
df_bristol_2025_09_26_neighbourhoods["compactness"] = (
    df_bristol_2025_09_26_neighbourhoods["perimeter_m"] ** 2
) / (4 * 3.1416 * (df_bristol_2025_09_26_neighbourhoods["area_sqkm"] * 1e6))
```

```python
BRISTOL_CITY_CENTER = (51.4545, -2.5879)

df_bristol_2025_09_26_neighbourhoods["dist_to_center_km"] = df_bristol_2025_09_26_neighbourhoods.apply(
    lambda row: geodesic(
        (row.centroid_lat, row.centroid_lon),
        BRISTOL_CITY_CENTER
    ).km,
    axis=1
)
```
