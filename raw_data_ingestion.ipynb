```python
url_bristol_listings_2025_09_26 = 'https://data.insideairbnb.com/united-kingdom/england/bristol/2025-09-26/data/listings.csv.gz'
url_bristol_calendar_2025_09_26 = 'https://data.insideairbnb.com/united-kingdom/england/bristol/2025-09-26/data/calendar.csv.gz'
url_bristol_reviews_2025_09_26 = 'https://data.insideairbnb.com/united-kingdom/england/bristol/2025-09-26/data/reviews.csv.gz'
url_bristol_neighbourhoods_2025_09_26 = 'https://data.insideairbnb.com/united-kingdom/england/bristol/2025-09-26/visualisations/neighbourhoods.csv'

url_bristol_listings_2025_06_24 = 'https://data.insideairbnb.com/united-kingdom/england/bristol/2025-06-24/data/listings.csv.gz'

url_bristol_listings_2025_03_19 = 'https://data.insideairbnb.com/united-kingdom/england/bristol/2025-03-19/data/listings.csv.gz'

url_edinburgh_listings_2025_09_21 = 'https://data.insideairbnb.com/united-kingdom/scotland/edinburgh/2025-09-21/data/listings.csv.gz'

url_london_listings_2025_09_14 = 'https://data.insideairbnb.com/united-kingdom/england/london/2025-09-14/data/listings.csv.gz'
```

```python
df_bristol_2025_09_26_listings = pd.read_csv(url_bristol_listings_2025_09_26)
df_bristol_2025_09_26_listings['city'] = 'bristol'
df_bristol_2025_09_26_listings['extraction_date'] = '2025-09-26'

df_bristol_2025_06_24_listings = pd.read_csv(url_bristol_listings_2025_06_24)
df_bristol_2025_06_24_listings['city'] = 'bristol'
df_bristol_2025_06_24_listings['extraction_date'] = '2025-06-24'

df_bristol_2025_03_19_listings = pd.read_csv(url_bristol_listings_2025_03_19)
df_bristol_2025_03_19_listings['city'] = 'bristol'
df_bristol_2025_03_19_listings['extraction_date'] = '2025-03-19'

df_edinburgh_2025_09_21_listings = pd.read_csv(url_edinburgh_listings_2025_09_21)
df_edinburgh_2025_09_21_listings['city'] = 'edinburgh'
df_edinburgh_2025_09_21_listings['extraction_date'] = '2025-09-21'

df_london_2025_09_14_listings = pd.read_csv(url_london_listings_2025_09_14)
df_london_2025_09_14_listings['city'] = 'london'
df_london_2025_09_14_listings['extraction_date'] = '2025-09-14'
```

```python
def compare_df_dtypes(df1, df2, df3, names=["DF1", "DF2", "DF3"]):
    # Union of all columns
    all_cols = sorted(set(df1.columns) | set(df2.columns) | set(df3.columns))

    dtype_comparison = pd.DataFrame({
        names[0]: df1.reindex(columns=all_cols).dtypes.astype(str),
        names[1]: df2.reindex(columns=all_cols).dtypes.astype(str),
        names[2]: df3.reindex(columns=all_cols).dtypes.astype(str),
    })

    # Flag rows where not all dtypes are identical
    differences = dtype_comparison[
        dtype_comparison.nunique(axis=1) > 1
    ]

    if differences.empty:
        print("✅ All columns have matching dtypes across all DataFrames.")
        return None
    else:
        print("❌ Found dtype differences in the following columns:")
        return differences

important_dtypes = {
    'host_listings_count': 'float64',
    'host_total_listings_count': 'float64',
    'license':'str',
    'minimum_minimum_nights':'float64',
    'maximum_minimum_nights':'float64',
    'minimum_maximum_nights':'float64',
    'maximum_maximum_nights':'float64'
}

df_bristol_2025_09_26_listings = df_bristol_2025_09_26_listings.astype({k: v for k, v in important_dtypes.items() if k in df_bristol_2025_09_26_listings.columns})
df_bristol_2025_06_24_listings = df_bristol_2025_06_24_listings.astype({k: v for k, v in important_dtypes.items() if k in df_bristol_2025_09_26_listings.columns})
df_bristol_2025_03_19_listings = df_bristol_2025_03_19_listings.astype({k: v for k, v in important_dtypes.items() if k in df_bristol_2025_09_26_listings.columns})

df_london_2025_09_14_listings = df_london_2025_09_14_listings.astype({k: v for k, v in important_dtypes.items() if k in df_london_2025_09_14_listings.columns})
df_edinburgh_2025_09_21_listings = df_edinburgh_2025_09_21_listings.astype({k: v for k, v in important_dtypes.items() if k in df_edinburgh_2025_09_21_listings.columns})
```

```python
diffs = compare_df_dtypes(df_london_2025_09_14_listings, df_edinburgh_2025_09_21_listings, df_bristol_2025_09_26_listings)
diffs
```
